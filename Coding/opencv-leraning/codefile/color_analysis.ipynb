```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```python
img=  cv2.imread("123.jpg")
chanles = cv2.split(img)
coloes = ("b","g","r")
plt.figure()
plt.title("DJH")
plt.xlabel("bins")
plt.ylabel("of Pixels")
for chan, color in zip(chanles, coloes):
    hist = cv2.calcHist([chan], [0], None, [255], [0,255])
    plt.plot(hist, color=color)
    plt.xlim([0,255])
```
