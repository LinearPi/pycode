```python
cap = cv2.VideoCapture(1)
i = 1
while(1): 
    _, frame = cap.read()
    fps =cap.get(cv2.CAP_PROP_FPS)
    print(fps)
#     gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    # 人像图像先减少一半的数据
    # 利用均值直方图增强图片的识别的内容
    gray = cv2.resize(frame, dsize=(0,0), fx =0.5, fy=0.5)
#     gray = cv2.equalizeHist(gray)
#     gray = frame
    
    #人脸分类器
    detector = dlib.get_frontal_face_detector()
    # 获取人脸检测器
    predictor = dlib.shape_predictor(r"F:\PYcode\coding\faceinfo\dlib_shapedata\shape_predictor_68_face_landmarks.dat")
    
    dets = detector(gray, 1)
    for face in dets:
        shape = predictor(gray, face)  
        # 寻找人脸的68个标定点
        # 遍历所有点，打印出其坐标，并圈出来
        face_list = []
        for pt in shape.parts():
            pt_pos = (pt.x, pt.y)
            face_list.append(pt_pos)
            cv2.circle(gray, pt_pos, 2, (0, 255, 0), 1)
        print(face_list[0])
        print(face_list[1])

        cv2.polylines(gray, [np.array(face_list[0:17])], False, (0,255,255), 1)
        cv2.polylines(gray, [np.array(face_list[17:22])], False, (0,0,255), 2)
        cv2.polylines(gray, [np.array(face_list[22:27])], False, (0,0,255), 2)
 
        cv2.polylines(gray, [np.array(face_list[36:42])], True, (0,255,255), 1)
        cv2.polylines(gray, [np.array(face_list[42:48])], True, (0,255,255), 1)

#     cv2.imshow('frame',frame) 
    cv2.imshow('gray',gray) 
    if i % 5 ==0:
        cv2.imwrite(f"cam_img\\{i}.jpg", frame)
    print(i)
    i += 1
    
    if cv2.waitKey(1) == ord('q'):
        break
cv2.destroyAllWindows()
cap.release()
```
